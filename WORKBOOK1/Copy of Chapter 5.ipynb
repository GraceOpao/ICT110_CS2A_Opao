## CHALLENGES

Challeng 1

```python
total = 0
count = 0

while True:
    num = input("Enter a number: ")
    if num.lower() == 'done':
        break

    try:
        number = float(num)
        total += number
        count += 1
        print("Total:", total)
        print("Count:", count)
        print("Average:", total / count)
    except ValueError:
        print("Invalid input")

if count == 0:
    print("No valid numbers were entered.")
```

#### Challenge 2  

```python
number = []

while True:
    try:
      number_input = input("Enter a number: ")
      if number_input == 'done':
        break
      num = float(number_input)
      number.append(num)
    except ValueError:
      print("Invalid input")

if number:
    maximum = max(number)
    minimum = min(number)
    print("Maximum:", maximum)
    print("Minimum:", minimum)

else:
    print("No numbers entered.")
```
